```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Define el sistema de ecuaciones diferenciales
def sistema(y, t):
    # y[0] y y[1] son las variables de estado
    dydt = [0.5*y[0]-y[1]-0.5*(y[0]*y[0]*y[0]+y[1]*y[1]*y[0]), y[0]+0.5*y[1]-0.5*(y[1]*y[1]*y[1]+y[0]*y[0]*y[1])]
    return dydt

# Condiciones iniciales
tam=[0.0000001,0.01,1,2,-0.0000001,-0.01,-1,-2,1.5,-1.5]

#Definici칩n de variables y de detalles en la gr치fica
y0=[]
sol=[]
j=0
plt.figure(figsize=(6, 6))
plt.title('Plano Fase de un Sistema de Ecuaciones Diferenciales')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(y=0,color='k')
plt.axvline(x=0,color='k')
plt.legend()
plt.grid(True)

for i in tam:
  y0.append([i,i])
# Tiempo de integraci칩n
  t = np.linspace(0, 100, 1000)
# Resuelve el sistema de ecuaciones diferenciales
  sol.append(odeint(sistema, y0[j], t))
# Grafica el plano fase
  plt.plot(sol[j][:, 0], sol[j][:, 1], label='Plano fase')
  j=j+1

#Mostrar gr치fica
plt.show()
```
