```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
#ingresar los datos
datos=pd.read_excel('https://github.com/artunam/Projects/raw/main/datos.xlsx')
#comienza el código
ma = datos['muestra'].tolist()
n=len(ma)
ma.sort()
m = pd.unique(ma)
k=m[1:]
aux=k[len(k)-1]+1
k=np.append(k,aux)
func = []
funcdesf = [0]
for a in m:
    func.append(np.sum(ma<=a)/n)
    funcdesf.append(np.sum(ma<=a)/n)
plt.grid()
plt.scatter(k,func,marker='o', facecolors='none', edgecolors='blue')
plt.scatter(m,func,marker='o',  edgecolors='blue')
for i in range(len(m)):
    plt.hlines(y=func[i], xmin=m[i], xmax=k[i], color='blue', linestyle='-')
plt.scatter(k[len(k)-1],1,marker='o',edgecolors='white')
plt.scatter(k[len(k)-1],1,marker='>',facecolors='blue',edgecolors='blue')
plt.scatter(m[0],0,marker='o', facecolors='none', edgecolors='blue')
aux2=m[0]-1
plt.hlines(y=0, xmin=aux2, xmax=m[0], color='blue', linestyle='-')
plt.scatter(m[0]-1,0,marker='<', edgecolors='blue')
plt.xticks(range(int(aux2), int(aux)+1, 1))
plt.xlabel('x')
plt.ylabel('Fn(x)')
plt.title('Función de distribución empírica')
plt.show()
```
