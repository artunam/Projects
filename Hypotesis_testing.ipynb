```python
from scipy import stats
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
%matplotlib inline
def dist_n(data):
        conteo, cortes = np.histogram(data, bins= 200)
        densidad = conteo/sum(conteo)
        acumulada = np.cumsum(densidad)
        dominio = np.unique(sorted(data)).tolist()
        rango = np.unique(acumulada).tolist()
        datos_de_salida = pd.DataFrame(list(zip(dominio,rango)), columns=['X','Fx'])
        return(datos_de_salida)
```

```python
#Ejercicio 1 CON KRUSKAL-WALLES
####DATA######
alpha=0.01
Bajo=[76,76,85,86,89,95,96,102,108]
Medio=[78,82,92,97,111]
Alto=[75,76,79,80,96]
muestras=[Bajo,Medio,Alto]
plt.hist(Bajo)
plt.hist(Medio)
plt.hist(Alto)
resultado=stats.kruskal(Bajo,Medio,Alto)
##############
plt.show()
k=len(muestras)
Valorcritico=stats.chi2.ppf(alpha,k-1)
H=resultado.statistic
respuesta="NO RECHAZAR"
mayor="menor"
diferentes="iguales"
hay="no hay"
if H>Valorcritico:
  respuesta="RECHAZAR"
  mayor="mayor"
  diferentes="diferentes"
  hay="hay"
print("Como el estadístico de Kruskal-Walles da ",H)
print("y es ",mayor," que ",Valorcritico," el valor del cuantil de una")
print("chi^2 con ", k-1," grados de libertad y tamaño ", alpha)
print("se determina ",respuesta, " la hipótesis nula.")
print("Es decir, hay evidencia suficiente para decir que")
print("las medianas de los grupos son ", diferentes,".")
print("Por tanto, ",hay," diferencia entre grupos.")


#Como el estadistico de Kruskal-Walles es mayor que el cuantil de la chi^2(2,0.01), esto sugiere que hay
#evidencia para RECHAZAR la hipótesis nula, es decir, hay evidencia
#suficiente para afirmar que las medianas de los grupos son diferentes
#es decir que hay diferencia entre los grupos
```

```python
#Ejercicio 4 CON WILCOXON
####DATA######
alpha=0.05
A=[12.4,12.7,12.9,13.3,14.2,14.3,14.3,14.8,14.8,15.3]
B=[10.7,12.2,12.8,13.9,14.1,14.3,14.6,15.6,16.8,18.1]
plt.hist(A)
plt.hist(B)
muestras=[A,B]
resultado=stats.wilcoxon(A,B)
##############
plt.show()
k=len(muestras)
Valorcritico=alpha
H=resultado.pvalue
respuesta="RECHAZAR"
mayor="menor"
diferentes="grandes"
hay="hay"
if H>Valorcritico:
  respuesta="NO RECHAZAR"
  mayor="mayor"
  diferentes="ínfimas"
  hay="no hay"
print("Como el pvalue de Wilcoxon da ",H)
print("y es ",mayor," que el valor alpha=",Valorcritico," del tamaño de la prueba.")
print("Se determina ",respuesta, " la hipótesis nula.")
print("Es decir, hay evidencia suficiente para decir que")
print("las distancias entre los grupos son ", diferentes,".")
print("Por tanto, ",hay," diferencia entre grupos.")

#Como el pvalue de Wilcoxon da  0.3135132343764381
#y es  mayor  que el valor alpha= 0.05  del tamaño de la prueba.
#Se determina  NO RECHAZAR  la hipótesis nula.
#Es decir, hay evidencia suficiente para decir que
#las distancias entre los grupos son ínfimas .
#Por tanto,  no hay  diferencia entre grupos.
```

```python
def wilcoxon(X, Y):
    # Calcular las diferencias entre las observaciones de cada par
    D = [round(x - y,10) for x, y in zip(X, Y)]

    # Obtener las diferencias absolutas y ordenarlas
    absD = sorted(abs(diff) for diff in D if diff != 0)

    # Inicializar el diccionario de rangos con todas las diferencias posibles
    rangos = {diff: 0.5 for diff in absD}

    # Calcular los rangos de las diferencias absolutas considerando empates
    rango_actual = 1
    i = 0
    while i < len(absD):
        j = i + 1
        while j < len(absD) and absD[j] == absD[i]:
            j += 1
        if j - i > 1:  # Hay empates
            suma_rango = sum(rango_actual + k for k in range(j - i))
            rango = suma_rango / (j - i)  # Dividir entre el número de elementos que empatan
        else:
            rango = rango_actual
        for k in range(i, j):
            rangos[absD[k]] = rango
        rango_actual += j - i
        i = j

    # Calcular la suma de los rangos de las diferencias positivas y negativas
    Tmas = sum(rangos[diff] for diff in D if diff > 0)
    Tmenos = sum(rangos[-diff] for diff in D if diff < 0)
    T=min(Tmas,Tmenos)
    return Tmas, Tmenos,T

X = [12,5.2,3.7,2.3,3.5,1,2.4,3.7,4.5,0]
Y = [12,5,4,2.5,3.3,1,3,3.4,4.4,0.15]

Tmas, Tmenos,T = wilcoxon(X, Y)

print("Suma de los rangos de las diferencias positivas:", Tmas)
print("Suma de los rangos de las diferencias negativas:", Tmenos)
print("El estadístico de Wilcoxon es:", T)
```

```python
from scipy.stats import norm

def wilcoxon_critical_value(n, alpha):
    # Calcular la media y la desviación estándar de la distribución normal estándar
    mu = n * (n + 1) / 4
    sigma = (n * (n + 1) * (2 * n + 1) / 24) ** 0.5

    # Calcular el valor crítico usando la distribución normal estándar
    z_critical = norm.ppf(1 - alpha/2)

    # Calcular el valor crítico ajustado para muestras pequeñas (n <= 20)
    if n <= 20:
        z_critical_adjusted = z_critical - (0.5 / n)
        return mu - z_critical_adjusted * sigma, mu + z_critical_adjusted * sigma
    else:
        return mu - z_critical * sigma, mu + z_critical * sigma

# Ejemplo de uso
n = 8  # Tamaño de la muestra
alpha = 0.05  # Nivel de significancia

lower_critical, upper_critical = wilcoxon_critical_value(n, alpha)
print("Valor crítico inferior:", lower_critical)
print("Valor crítico superior:", upper_critical)
```

```python
from scipy.stats import norm

def wilcoxon_critico(n, alpha):
    # Calcular la media y la desviación estándar de la distribución normal estándar
    mu = n * (n + 1) / 4
    sigma = (n * (n + 1) * (2 * n + 1) / 24) ** 0.5

    # Calcular el valor crítico usando la distribución normal estándar
    z_critico = norm.ppf(1 - alpha/2)

    # Calcular el valor crítico ajustado para muestras pequeñas (n <= 20)
    if n <= 20:
        z_critico_ajustado = z_critico - (0.5 / n)
        return mu - z_critico_ajustado * sigma, mu + z_critico_ajustado * sigma
    else:
        return mu - z_critico * sigma, mu + z_critico * sigma

# Ejemplo de uso
n = 7  # Tamaño de la muestra
alpha = 0.05  # Nivel de significancia

lower_critical, upper_critical = wilcoxon_critical_value(n, alpha)
print("Valor crítico inferior:", lower_critical)
print("Valor crítico superior:", upper_critical)
```

```python
def mann_whitney_ranks(X, Y):
    # Combinar y ordenar las observaciones de ambas muestras
    combined = sorted(list(zip(X, ['X']*len(X))) + list(zip(Y, ['Y']*len(Y))))

    # Inicializar los rangos y la suma de los rangos
    ranks = {}
    sum_ranks_X = 0
    sum_ranks_Y = 0
    current_rank = 1

    # Asignar rangos y calcular la suma de los rangos
    for i, (value, sample) in enumerate(combined):
        if i > 0 and combined[i-1][0] == value:
            ranks[value] = ranks[combined[i-1][0]]
        else:
            ranks[value] = current_rank
        if sample == 'X':
            sum_ranks_X += ranks[value]
        else:
            sum_ranks_Y += ranks[value]
        current_rank += 1

    return sum_ranks_X, sum_ranks_Y

# Ejemplo de uso
X = [8,7,8,6,6.5,6]
Y=[5,3.5,5.5,5,3.5]

sum_ranks_X, sum_ranks_Y = mann_whitney_ranks(X, Y)
print("Suma de los rangos de la muestra X:", sum_ranks_X)
print("Suma de los rangos de la muestra Y:", sum_ranks_Y)
```
